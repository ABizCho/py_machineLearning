1. define convertCelsius()

```python
import math

# input is celsius value.
# output is float type fahrenheit value. (print it)
def convertCelsius():
    celsius = input("Enter the tamperature in degrees Celsius: ")
    fahrenheit = float(celsius) * (9 / 5) + 32
    print(f"{fahrenheit} degrees Fahrenheit")
    return None
```

3. define quadraticRoot()

```python
def quadraticRoot():
    a = int(input("Enter the a value")) # According to instruction, integer only 
    b = int(input("Enter the b value"))
    c = int(input("Enter the c value"))

    result_x1 = (-b + (b ** 2 - 4 * a * c) ** 0.5) / 2 * a
    result_x2 = (-b - (b ** 2 - 4 * a * c) ** 0.5) / 2 * a
    
    print(f"The solutions include {result_x1} and {result_x2}")

    return None
```

4. define calcDistance()

```python
def calcDistance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2) # return inline (** is a math operater; (value)^2)
```
