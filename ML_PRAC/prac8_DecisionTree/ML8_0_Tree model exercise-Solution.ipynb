Now let's create the model data for the logistic regression model. Suppose that you are using the following set of variables as predictors: ['ln_goal','ln_duration','facebook_url', 'imdb_url', 'twitter_url', 'youtube_url', 'website_url', 'featured', 'enable_drcc', 'enable_payp','all_or_nothing','funding_started_at_DOW_1.0','funding_started_at_DOW_2.0','funding_started_at_DOW_3.0','funding_started_at_DOW_4.0','funding_started_at_DOW_5.0','funding_started_at_DOW_6.0']

Note that there are missing values. Drop all the observations with any missing values among the set of predictors. After dropping them, specify your target variable and predictors. 

```python
#drop the observations that have missing values 
X_list=['ln_goal','ln_duration','facebook_url', 'imdb_url', 'twitter_url', 
        'youtube_url', 'website_url', 'featured', 'enable_drcc', 'enable_payp',
        'all_or_nothing','funding_started_at_DOW_1.0','funding_started_at_DOW_2.0',
        'funding_started_at_DOW_3.0','funding_started_at_DOW_4.0',
        'funding_started_at_DOW_5.0','funding_started_at_DOW_6.0'] #we use 'ln_goal' and 'ln_duration' instead of 'goal'and'duration'
new_crowdfunding=crowdfunding.dropna(axis=0,how='any',subset=X_list) #drop all rows that have any NaN values

#choose the dependent variable and independent variables 
X = new_crowdfunding[X_list]
y = new_crowdfunding['win']
```

```python
X.head()
```

Great! Our data is ready for our model!

## Build Random Forest model and evaluate the performance

Break data into training and test set.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)

rfc_pred = rfc.predict(X_test)

print(confusion_matrix(y_test,rfc_pred))

print(classification_report(y_test,rfc_pred))
```

```python
pd.DataFrame(rfc.feature_importances_, X_train.columns, columns=['Feature Importance'])
```

## Comparing four different machine learning models

Now let's compare multiple machine learning algorithms. We will compare 4 algorithms - Logistic Regression, Random Forest, KNN (K-nearest neighbor), and Naive Bayes.

#### try four different ML models and get an accuracy score for each of them. 

```python
# import different algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# get accuracy score
from sklearn.metrics import accuracy_score

# prepare models
models=[]
models.append(('LR',LogisticRegression()))
models.append(('RFC',RandomForestClassifier()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('NB',GaussianNB()))

# evaluate each model in turn
for name, model in models:
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    ac_score=accuracy_score(y_test,predictions)
    msg="%s: %f" % (name, ac_score)
    print(msg)  
```

The logistic regression model is best in terms of the accuracy. 

#### try four different ML models and get an f1-score score for each of them. 

```python
# get f1 score
from sklearn.metrics import f1_score

# evaluate each model in turn
for name, model in models:
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    f_score=f1_score(y_test,predictions)
    msg="%s: %f" % (name, f_score)
    print(msg)
```
