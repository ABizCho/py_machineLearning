## What is the most highly star-rated laptop brand?

```python
print('Total Mean of Star Rating :', df_laptop['star_rating'].mean() , '\n')

print('The highest star-rated brand : ',df_laptop.groupby('brand')['star_rating'].mean().sort_values(ascending=False).max(),'\n')
print('The lowest star-rated brand : ',df_laptop.groupby('brand')['star_rating'].mean().sort_values(ascending=False).min(),'\n')

print(df_laptop.groupby('brand')['star_rating'].mean().sort_values(ascending=False))
```
